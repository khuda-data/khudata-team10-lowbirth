```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 & 마이너스 깨짐 방지
mpl.rcParams["font.family"] = "Malgun Gothic"
mpl.rcParams["axes.unicode_minus"] = False
```

```python
df = pd.read_excel("b_sweep_results.xlsx")
```

```python
# 선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df['b'], df['total_rate'], label='혼인율', color='#25228F')

# 그래프 꾸미기
plt.xlabel('동질혼 선호 강도')
plt.ylabel('혼인율')
plt.title('동질혼 선호 강도에 따른 혼인율 변화')
plt.legend(fontsize=15)
# plt.grid(True)
plt.tight_layout()
plt.show()
```
